```python
import docx
from owlready2 import *
from unidecode import unidecode


# Create a new ontology
onto = get_ontology("http://example.org/appendix_ontology.owl")

# Define the class "Ngành nghề kinh doanh có điều kiện"
with onto:
	class ngành_nghề_kinh_doanh_có_điều_kiện(Thing):
		pass

# Save the ontology
onto.save("appendix_ontology.owl")

doc = docx.Document("67_2014_QH13_259729.docx")
table = doc.tables[-1]

for row in table.rows[1:]:
	cell = row.cells[-1]
	text = (cell.text).lower()
	# Add underscores between words
	text = '_'.join(text.split())

	# Add the text to the ontology as a subclass of "Ngành nghề kinh doanh có điều kiện"
	with onto:
		class_name = types.new_class(text, (ngành_nghề_kinh_doanh_có_điều_kiện,))
		print(class_name)
	print(cell.text)

# Save the ontology
onto.save("appendix_ontology.owl")
```
